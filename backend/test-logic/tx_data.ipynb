```python
import numpy as np

credit_card_names = ["U.S. Bank Cash Plus Visa Signature Card", "American Express Gold Card", "Chase Sapphire Reserve"]

def getPointPerCard(credit_card_names):
    points_per_card_per_year = {}
    for name in credit_card_names:

        points_dict = map_df[name][5]["pointsPerDollar"]
        card_name = name
        
        for (year, category), amount in grouped_df.items():
            if amount <= 0:  # Skip negative amounts (e.g., refunds)
                continue
            
            points_earned = 0
            # Calculate points earned:
            points_earned += amount * points_dict.get(category, points_dict.get("other", 0))
            
            if year not in points_per_card_per_year:
                points_per_card_per_year[year] = {}
            
            if card_name not in points_per_card_per_year[year]:
                points_per_card_per_year[year][card_name] = 0
            
            points_per_card_per_year[year][card_name] += points_earned
    return points_per_card_per_year

# Print or analyze the results
points_per_card_per_year = getPointPerCard(credit_card_names)
for year, card_points in points_per_card_per_year.items():
    print(f"Year: {year}")
    for card_name, points in card_points.items():
        print(f"  {card_name}: {points:.2f} points")
```

```python
import numpy as np

credit_card_names = ["U.S. Bank Cash Plus Visa Signature Card", "American Express Gold Card", "Chase Sapphire Reserve"]

def getTotalCombination(credit_card_names):
    best_card_per_category_per_year = {}
    for name in credit_card_names:
        # Adjusted access logic for map_df based on how your data is structured
        points_dict = map_df[name][5]["pointsPerDollar"]
        card_name = name
        
        for (year, category), amount in grouped_df.items():
            if amount <= 0:  # Skip negative amounts (e.g., refunds)
                continue
            
            points_earned = amount * points_dict.get(category, points_dict.get("other", 0))
            
            if year not in best_card_per_category_per_year:
                best_card_per_category_per_year[year] = {}
            
            # Check if the current card provides more points for this category
            if (category not in best_card_per_category_per_year[year] or
                points_earned > best_card_per_category_per_year[year][category]['points']):
                
                best_card_per_category_per_year[year][category] = {
                    'card_name': card_name,
                    'points': points_earned
                }

    # Summarize the total points for each year
    total_points_per_year = {}

    for year, categories in best_card_per_category_per_year.items():
        total_points = sum(details['points'] for details in categories.values())
        total_points_per_year[year] = total_points

    return best_card_per_category_per_year, total_points_per_year

# Get results
best_card_per_category_per_year, total_points_per_year = getTotalCombination(credit_card_names)

# Print detailed results
for year, categories in best_card_per_category_per_year.items():
    print(f"Year: {year}")
    for category, details in categories.items():
        print(f"  Category: {category} - Best Card: {details['card_name']} with {details['points']:.2f} points")
    print(f"  Total Points for {year}: {total_points_per_year[year]:.2f} points\n")
```

```python
def analyze_credit_card_points(grouped_df, credit_cards, map_df):
    """
    Analyzes credit card points based on transaction data and card rewards mapping.
    
    Args:
        grouped_df: DataFrame with transaction amounts grouped by year and category
        credit_cards: List of credit card names
        map_df: Dictionary containing points per dollar mapping for each card
    """
    best_card_per_category_per_year = {}
    total_points_per_year = defaultdict(float)
    
    # First pass: Calculate best card for each category and year
    for (year, category), amount in grouped_df.items():
        if amount <= 0:  # Skip negative amounts
            continue
            
        best_points = 0
        best_card = None
        
        # Check each card to find the best points earned
        for card in credit_cards:
            points_dict = map_df[card][5]["pointsPerDollar"]
            points_multiplier = points_dict.get(category, points_dict.get("other", 0))
            points_earned = amount * points_multiplier
            
            if points_earned > best_points:
                best_points = points_earned
                best_card = card
        
        # Store the best card and points for this category
        if year not in best_card_per_category_per_year:
            best_card_per_category_per_year[year] = {}
            
        best_card_per_category_per_year[year][category] = {
            'card_name': best_card,
            'points': best_points,
            'spend': amount
        }
        
        total_points_per_year[year] += best_points
    
    # Print results
    for year in sorted(best_card_per_category_per_year.keys()):
        print(f"\n{'='*80}")
        print(f"Year: {year}")
        print('='*80)
        print(f"{'Category':<20} {'Best Card':<35} {'Spend':<15} {'Points':<15}")
        print('-'*80)
        
        categories = best_card_per_category_per_year[year]
        for category, details in sorted(categories.items()):
            print(f"{category:<20} "
                  f"{details['card_name']:<35} "
                  f"${details['spend']:>10,.2f}   "
                  f"{details['points']:>10,.2f}")
        
        print('-'*80)
        print(f"Total Points for {year}:{' '*44}    {total_points_per_year[year]:>10,.2f}")

# Example usage
from collections import defaultdict

analyze_credit_card_points(grouped_df, credit_card_names, map_df)
```
